Придумайте матрицы, которые обладают следующими свойствами:

```python
import numpy as np
import math
```

```python
# a) диагональная, ортогональная, но не единичная(!):
a = np.array([ [-1,0,0],[0,-1,0],[0,0,1] ])
print(a)
# диагональная матрица, элементы диагонали которой равны либо +1 либо -1, является ортогональной
```

```python
# b) верхнетреугольная, ортогональная:

b = np.array([ [1,0,0],[0,1,0],[0,0,-1] ])
print(b)

# у ортогональной матрицы сумма квадратов элементов любой строки равна 1, 
# сумма произведений соответствующих элементов любых разных строк равна 0. Аналогичное свойство справедливо для столбцов.

# диагональная матрица - это пример матрицы, которая является одновременно верхне- и нижнетреугольной.
```

```python
# c) нижнетреугольная, ортогональная:

c = np.array([ [1,0],[0,-1] ])
print(c)
```

```python
# d) матрица, у которой определитель = 1, и все элементы не равны 0:

theta = 30 * math.pi/180 # угол 30 градусов в радианах
d = np.array([
    [math.cos(theta), -math.sin(theta)],
    [math.sin(theta), math.cos(theta)]
]) # матрица поворота в двумерном пространстве

print(d)
np.linalg.det(d)
```

```python
# e) матрица (3x3), у которой определитель = 1:

e = np.array([ [0.2,10,-7.5],[0,2,-30],[0,0,2.5] ])

print(e)
np.linalg.det(e)

# определитель верхне- (или нижне-) треугольной матрицы равен произведению диагональных элементов.
```

### Задание 2:

Найдите собственные значения и собственные вектора матриц (необходимо решение на numpy и честное решение на бумаге:)

```python
m = np.array([[2, 2],
              [1, 3]])
print(m)
```

```python
lmd, u = np.linalg.eig(m)

for num in range(len(lmd)):
    print(f'Собственное значение: {lmd[num]}\nсоответствующий ему собственный вектор: {u[:,num]}') 
```

```python
# Проверим равенство np.dot(m[:,:], u[:,i]) = lmd[i] * u[:,i], для всех i из интервала [0, ..., N-1], где N - размер матрицы

np.dot(m[:,:], u[:,0]) == lmd[0] * u[:,0]
np.dot(m[:,:], u[:,0]) - lmd[0] * u[:,0] # погрешность

np.dot(m[:,:], u[:,1]) == lmd[1] * u[:,1]
```

Найдите спектральное разложение матрицы (честное решение на бумаге и на numpy)

```python
m = np.array([[1, 2],
              [2,3]])
print(m)
```

```python
print(m == m.transpose()) 

# Матрица m является симметричной, поэтому для спектрального разложения использую формулу u.dot(np.diag(lmd)).dot(u.T)
# u - ортогональная матрица со столбцами из СВ матрицы m
# np.diag(lmd) - диагональная матрица, на главной диагонали которой СЗ матрицы m
```

```python
lmd, u = np.linalg.eig(m)

for num in range(len(lmd)):
    print(f'Собственное значение: {lmd[num]}\nсоответствующий ему собственный вектор: {u[:,num]}') 
```

```python
u
np.diag(lmd)
u.T
```

```python
np.linalg.inv(u) == u.T # проверка ортогональности матрицы
np.linalg.inv(u) - u.T # погрешность
```

### Задание 4:

Для примера из лекции постройте график зависимости oob_score RandomForestClassifier от количества компонент при сжатии с помощью SVD

```python
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams

from sklearn.datasets import load_digits
from sklearn.decomposition import TruncatedSVD
from sklearn.ensemble import RandomForestClassifier
```

```python
# Датафрейм для результатов
results = pd.DataFrame(columns=['n_components', 'oob_score_decr', 'oob_score_incr'])

n_components = 64
for idx in range(32):
    results.loc[idx] = [n_components,0,0]
    n_components-=2

results.head()
```

```python
X, y = load_digits(return_X_y=True)

X.shape
y.shape
```

```python
image = X[8]
image = image.reshape((8, 8))
plt.matshow(image, cmap = 'gray')
plt.show()
```

```python
rf_original = RandomForestClassifier(oob_score=True)
rf_original.fit(X, y)

results.oob_score_decr[0], results.oob_score_incr[0] = rf_original.oob_score_, rf_original.oob_score_

results.head()
```

```python
n_components_decr=62
for idx in range(1,32):
    svd = TruncatedSVD(n_components=n_components_decr)
    X_decr = svd.fit_transform(X)
    
    rf_decr = RandomForestClassifier(oob_score=True)
    rf_decr.fit(X_decr, y)
    
    results.oob_score_decr[idx] = rf_decr.oob_score_
    n_components_decr-=2
```

```python
results.head()
```

```python
n_components_incr=2
for idx in range(31,0,-1):
    svd = TruncatedSVD(n_components=n_components_incr)
    X_incr = svd.fit_transform(X)
    
    rf_incr = RandomForestClassifier(oob_score=True)
    rf_incr.fit(X_incr, y)
    
    results.oob_score_incr[idx] = rf_incr.oob_score_
    n_components_incr+=2
```

```python
results.head()
```

```python
def mean_error(row):
    return round((row['oob_score_decr']+row['oob_score_incr'])/2, 4)
```

```python
results['oob_score_mean'] = results.apply(mean_error, axis=1)
```

```python
results
```

```python
rcParams['figure.figsize'] = 12,7
plt.plot(results.n_components, results.oob_score_mean, color="blue", marker="o", markersize=3, linewidth=0.7, label='oob_score_mean')

plt.xlim(0,65)
plt.ylim(0.25,1)

plt.title('Зависимость oob_score RandomForestClassifier от количества компонент при сжатии с помощью SVD')
plt.grid()
plt.legend()
plt.xlabel('n_components')
plt.ylabel('oob_score')
```
