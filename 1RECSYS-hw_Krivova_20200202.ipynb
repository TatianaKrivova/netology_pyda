**Задание**. Постройте топ фильмов в категориях Action и Comedy.

Выбранные рейтинги:

**ТОП-20 фильмов с наибольшим количеством оценок 5.0**

**ТОП-20 фильмов с наибольшим показателем: количество оценок 4.5 и выше / количество лет с момента выпуска фильма**

```python
import pandas as pd
```

```python
links = pd.read_csv('links.csv')
links.shape # размер датасета
links.head() # справочник с ID фильмов

movies = pd.read_csv('movies.csv')
movies.shape
movies.head() # справочник фильмов и их жанров

ratings = pd.read_csv('ratings.csv')
ratings.shape
ratings.head() # логи оценок

tags = pd.read_csv('tags.csv')
tags.shape
tags.head() # логи комментариев
```

```python
def find_action(x):
    if 'action' in x['genres'].lower(): return 1
    else: return 0

def find_comedy(x):
    if 'comedy' in x['genres'].lower(): return 1
    else: return 0
```

```python
years = [y for y in range (1900, 2020)] # наполняю список годов

def production_year(line):
    '''Возвращает год выпуска из названия фильма. Аргумент (str): line.'''
    actual_year = [year for year in years if str(year) in line] # при обнаружении года в названии фильма добавляю в список
    
    if len(actual_year) == 1: return actual_year[0]
    elif len(actual_year) == 2: return actual_year[1]
    else: return 0
```

```python
movies['is_Action'] = movies.apply(find_action, axis=1)
movies['is_Comedy'] = movies.apply(find_comedy, axis=1)
movies['production_year'] = movies['title'].apply(production_year)

movies.head(10)
```

```python
movies.sort_values('production_year', ascending = False)
```

```python
movies.loc[9031, 'production_year'] = 2018
```

```python
movies['years_since_release'] = 2020 - movies['production_year']
```

```python
movies.sort_values('production_year', ascending = False)
```

```python
action_films = movies[ movies.is_Action == 1 ][ ['movieId', 'title', 'is_Action', 'years_since_release'] ]
comedy_films = movies[ movies.is_Comedy == 1 ][ ['movieId', 'title', 'is_Comedy', 'years_since_release'] ]

action_films.head()
comedy_films.head()
```

```python
films5 = ratings[ ratings.rating == 5.0 ] # фильмы, у которых есть оценка 5.0
films45 = ratings[ ratings.rating >= 4.5 ] # фильмы, у которых есть оценка 4.5 и выше
```

```python
films5.shape
films45.shape
```

```python
films5 = films5.groupby('movieId').count().sort_values('rating', ascending = False).reset_index()[['movieId','rating']]
films45 = films45.groupby('movieId').count().sort_values('rating', ascending = False).reset_index()[['movieId','rating']]

films5.head(20)
films45.head(20)
```

```python
action_films5 = action_films.merge(films5, on='movieId', how='left').sort_values('rating', ascending = False).reset_index()
comedy_films5 = comedy_films.merge(films5, on='movieId', how='left').sort_values('rating', ascending = False).reset_index()
```

```python
top20_action5 = action_films5[['title','rating']].loc[0:19]
top20_comedy5 = comedy_films5[['title','rating']].loc[0:19]
```

```python
top20_action5 # ТОП-20 фильмов с наибольшим количеством оценок 5.0 в категории Action
```

```python
top20_comedy5 # ТОП-20 фильмов с наибольшим количеством оценок 5.0 в категории Comedy
```

```python
action_films45 = action_films.merge(films45, on='movieId', how='left')
comedy_films45 = comedy_films.merge(films45, on='movieId', how='left')
```

```python
action_films45['new_rating'] = action_films45['rating'] / action_films45['years_since_release']
comedy_films45['new_rating'] = comedy_films45['rating'] / comedy_films45['years_since_release']
```

```python
action_films45.sort_values('new_rating', ascending = False, inplace=True)
comedy_films45.sort_values('new_rating', ascending = False, inplace=True)
```

```python
top20_action45 = action_films45.reset_index()[['title','new_rating']].loc[0:19]
top20_comedy45 = comedy_films45.reset_index()[['title','new_rating']].loc[0:19]
```
