```python
import numpy as np
import math

from sympy import *

from scipy.misc import derivative
```

```python
# с помощью sympy

x =  var('x')
func = x**5 + 4*sin(2*x)+cos(3*x+3) 
```

```python
deriv = func.diff(x)
```

```python
for rate in range(1,5):    
    if rate==3 or rate==4:
        print(f'{rate}-я производная:')
        deriv
        print(f'Значение в точке x=1:')
        deriv.evalf(subs={x: 1.0}, n=10)
    deriv = deriv.diff(x)
```

```python
# с помощью scipy

def f(i):
    return i**5 + 4*math.sin(2*i)+math.cos(3*i+3)
```

```python
derivative(f, 1.0, dx=0.01, n=3, order=61)
derivative(f, 1.0, dx=0.01, n=4, order=61)
```
