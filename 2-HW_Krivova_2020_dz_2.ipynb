Пример линейной регрессии - это применение полиномов на практике. В следующем примере дан датасет x, y, сгенерированный рандомно. Мы можем использовать линейную регрессию, чтобы предсказать значение переменной y от x. 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20) # random samples from a normal distribution: centre, width, output shape
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)
```

```python
plt.scatter(x, y, marker='o')

plt.title('Точечный график')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```python
X = np.array(x).reshape(-1,1)

model = LinearRegression() # Линейная регрессия
result = model.fit(X, y)

k = model.coef_
b = model.intercept_

print('slope, k:', k)
print('intercept, b:', b)
```

```python
pred_values = model.predict(X)
```

```python
plt.title('Линейная регрессия')

plt.plot(x, y, 'r.')
plt.plot(X,pred_values)

plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend(['исходные данные', 'полином 1ой степени'])

plt.show()
```

```python
# Метрики качества модели Линейной регрессии
print(f'mean_absolute_error: {mean_absolute_error(y, pred_values)}')
print(f'mean_squared_error: {mean_squared_error(y, pred_values)}')
print(f'r2_score: {r2_score(y, pred_values)}')
```

В данном примере найден полином первой степени. Как можно улучшить данную модель? Написать пример. 

```python
cubic = PolynomialFeatures(degree=3) # Улучшение модели за счет использования полинома третьей степени
X_cubic = cubic.fit_transform(X)
  
cubic.fit(X_cubic, y)

regr = LinearRegression()
model = regr.fit(X_cubic, y)
```
