```python
# Читаю файл ratings.csv по частям (часть = 1 млн строк), чтобы избежать ошибки памяти. Размер файла 724 MB, 27 млн строк.
# В датафрейм films5 добавляю фильмы, у которых есть оценка 5.0  

films5 = pd.concat([element[ element.rating == 5.0 ] for element in pd.read_csv('ratings.csv', chunksize=10**6)],\
                  ignore_index=True) # films5.shape --> (4071135, 4)
```

**Задание 2**. По данным файла *power.csv* посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
general_info = pd.read_csv('power.csv')
balic, categories = ['Latvia', 'Lithuania', 'Estonia'], [4, 12, 21]
```

```python
filtered_table = general_info[(general_info.country.isin(balic)) &\
                              (general_info.category.isin(categories)) &\
                              ((general_info.year >= 2005) & (general_info.year <= 2010)) &\
                              (general_info.quantity > 0)]
filtered_table['quantity'].sum()
```
