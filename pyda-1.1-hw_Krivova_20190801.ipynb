**Задание 3**. Разработать приложение для определения знака зодиака по дате рождения. Пример:

Введите месяц: март

Введите число: 6

Вывод:
Рыбы

```python
from datetime import date # из модуля datetime импортирую класс date для работы с датами
months = ['январь', 'февраль', 'март', 'апрель', 'май', 'июнь',\
          'июль', 'август', 'сентябрь', 'октябрь', 'ноябрь', 'декабрь'] # создаю список месяцев
leap_year = 2016 # високосный год

# создаю словарь зодиаков, в котором ключ - это дата окончания периода зодиака (включительно), 
# а значение - название соответствующего зодиака 
zodiac = {(date(leap_year,1,20)): 'Козерог',\
          (date(leap_year,2,19)): 'Водолей',\
          (date(leap_year,3,20)): 'Рыбы',\
          (date(leap_year,4,20)): 'Овен',\
          (date(leap_year,5,21)): 'Телец',\
          (date(leap_year,6,21)): 'Близнецы',\
          (date(leap_year,7,22)): 'Рак',\
          (date(leap_year,8,21)): 'Лев',\
          (date(leap_year,9,23)): 'Дева',\
          (date(leap_year,10,23)): 'Весы',\
          (date(leap_year,11,22)): 'Скорпион',\
          (date(leap_year,12,22)): 'Стрелец',\
          (date(leap_year+1,1,20)): 'Козерог'} 

# пользовательский ввод
current_mon = input('Введите месяц: ').lower().rstrip()
current_day = input('Введите число: ').rstrip()

# проверка: наличие введенного пользователем месяца в списке months и корректность введенного числа
if current_mon in months and current_day.isdigit():  
    current_mon = months.index(current_mon) + 1 # переменной current_mon присваивается номер соответствующего месяца
    try:
        # проходим по ключам словаря (верхним границам зодиаков) 
        # и находим первое значение, которое превысит или совпадет с проверяемой датой
        for key in zodiac:
            if key >= date(leap_year, current_mon, int(current_day)): 
                print(zodiac[key]) 
                break
             
    except ValueError as ve: # обработка ошибки: в году не существует введенной пользователем комбинации месяца и даты
        print('Такой даты не существует:', ve)
else: print ('Некорректный ввод')
```

**Задание 4**. Нужно разработать приложение для финансового планирования.
Приложение учитывает сколько уходит на ипотеку, "на жизнь" и сколько нужно отложить на пенсию.
Пользователь вводит:
- заработанную плату в месяц.
- сколько процентов от ЗП уходит на ипотеку.
- сколько процентов от ЗП уходит "на жизнь".
- сколько раз приходит премия в год.

Остальная часть заработанной платы откладывается на пенсию.

Также пользователю приходит премия в размере зарплаты, от которой половина уходит на отпуск, а вторая половина откладывается.

Программа должна учитывать сколько премий было в год.

Нужно вывести сколько денег тратит пользователь на ипотеку и сколько он накопит за год.

Пример:

Введите заработанную плату в месяц: 100000

Введите сколько процентов уходит на ипотеку: 30

Введите сколько процентов уходит на жизнь: 50

Введите количество премий за год: 2

Вывод:
На ипотеку было потрачено: 360000 рублей Было накоплено: 340000 рублей

```python
def fin_plan():
    user_finance = {}
    elements = ['зарплату в месяц', 'сколько процентов уходит на ипотеку',\
           'сколько процентов уходит на жизнь', 'количество премий за год']
    try:
        # пользовательский ввод и преобразование в десятичное число
        for element in elements:
            user_finance[element] = float(input('Введите '+ element +': ').rstrip())
        
        # проверка условия, что сумма процентов не превышает 100 
        if (user_finance['сколько процентов уходит на ипотеку'] + user_finance['сколько процентов уходит на жизнь']) >=0 and\
(user_finance['сколько процентов уходит на ипотеку'] + user_finance['сколько процентов уходит на жизнь']) <=100:
            
            mortgage = user_finance['зарплату в месяц']*user_finance['сколько процентов уходит на ипотеку']*12/100
            
            benefit = user_finance['зарплату в месяц']*12*\
(100 - user_finance['сколько процентов уходит на ипотеку'] - user_finance['сколько процентов уходит на жизнь'])/100 +\
user_finance['количество премий за год']*user_finance['зарплату в месяц']/2
            
            print(f'\nНа ипотеку было потрачено: {round(mortgage,2)} рублей. Было накоплено: {round(benefit,2)} рублей')
        
        else: print ('Некорректный ввод %')
                
    except ValueError as ve: # обработка ошибки: пользователь ввел не числа
        print('Ошибка ввода:', ve)

fin_plan() # вызов функции
```
