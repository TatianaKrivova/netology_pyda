**Задание 1**. Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам. Результат классификации запишите в столбец *class*.
        
    Оценка 2 и меньше - низкий рейтинг; оценка 4 и меньше - средний рейтинг; оценка 4.5 и 5 - высокий рейтинг.

```python
import pandas as pd

films = pd.read_csv('ratings.csv') # films.shape --> (100004, 4)
```

```python
def classify_rating(mark):
    '''Возвращает рейтинг фильма по его оценке. Аргумент (int, либо float): mark.'''
    if mark <= 2: return 'низкий рейтинг'
    elif mark <= 4: return 'средний рейтинг'
    elif mark <= 5: return 'высокий рейтинг'
    else:
        return 'undefined'
```

```python
films['class'] = films['rating'].apply(classify_rating)
films 
```

**Задание 2**. Используем файл *keywords.csv*. Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Результат классификации запишите в отдельный столбец region.

Т.е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'. Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

*geo_data = { 'Центр': ['москва', 'тула', 'ярославль'], 'Северо-Запад': ['петербург', 'псков', 'мурманск'], 'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }*.

```python
table = pd.read_csv('keywords.csv') # table.shape --> (100000, 2)
```

```python
def classify_region(line):
    '''Возвращает регион по вхождению города в запрос. Аргумент (str): line.'''
    
    geo_data = {'Центр': ['москва', 'тула', 'ярославль'],\
                'Северо-Запад': ['петербург', 'псков', 'мурманск'],\
                'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
    
    match = 0 # счетчик количества совпадений города в строке запроса
    region = [] # в этот список буду добавлять ключи словаря при обнаружении города в запросе
    
    for word in line.lower().strip().split(): # запрос привожу к нижнему регистру и разбиваю на слова
        for key, element in geo_data.items(): # прохожу по словарю и нахожу совпадения
            if word in element: 
                match += 1
                region.append(key)
    
    if match == 0: return 'undefined'
    elif match == 1 or (match == 2 and region[0] == region[1]): return region[0]
    elif match == 2 and region[0] != region[1]: return region[0], region[1]
    else: return region
    
# Проверка данных: Количество совпадений города(match) / в строках(keyword): 0 - 99406, 1 - 590, 2 - 4. 
    # т.е. в четырех запросах встречаются названия двух городов: москва ... петербург; ярославль ... ярославль
        # если города в запросе принадлежат к разным регионам - вывожу оба региона: Центр, Северо-Запад
        # если к одному региону - вывожу один регион: Центр
```

```python
table['region'] = table['keyword'].apply(classify_region)
table['region'].value_counts()
```

**Задание 3**. Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из материалов занятия. Т.е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже. При этом мы не будем затрагивать субъективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную *years* запишите список из всех годов с 1950 по **2017**.
2. Напишите функцию *production_year*, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
    - для каждой строки пройдите по всем годам списка years
    - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец *'year'*.
4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца *'year'* и отсортируйте результат по убыванию рейтинга.

```python
years = [y for y in range (1950, 2018)] # наполняю список годов
catalog = pd.read_csv('movies.csv') # catalog.shape --> (9125, 3)
films = pd.read_csv('ratings.csv') # films.shape --> (100004, 4)
```

```python
def production_year(line):
    '''Возвращает год выпуска из названия фильма. Аргумент (str): line.'''
    actual_year = [year for year in years if str(year) in line] # при обнаружении года в названии фильма добавляю в список
    
    if len(actual_year) == 1: return actual_year[0]
    elif len(actual_year) == 2: return actual_year[1]
    else: return 1900

# Проверка данных: Количество совпадений года в строках(title): 2 - 14. 
# т.е. в названии четырнадцати фильмов встречается два года. Годом выпуска в этом случае считаю последний найденный год.
```

```python
catalog['year'] = catalog['title'].apply(production_year)
result = films.merge(catalog, on='movieId', how='left')

result[ result.year != 1900 ][ ['rating', 'year'] ].groupby('year').mean().sort_values('rating', ascending = False)
```
