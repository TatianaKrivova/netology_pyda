этап 2: взять два атрибута и посчитать скалярное произведение двух векторов (2 дома), начертить 2 дома как вектора в двухмерном пространстве, и сравниваем графически и через скалярное произведение.


Возьмем атрибуты:
* DIS: This is the weighted distances to five Boston employment centers
* PTRATIO: This is the pupil-teacher ratio by town

```python
v1 = np.array([boston.data[257][7], boston.data[257][10]])
v2 = np.array([boston.data[355][7], boston.data[355][10]])

v1
v2
print(f'Скалярное произведение: {np.dot(v1, v2)}')
```

```python
rcParams['figure.figsize'] = 7,7

plt.scatter(v1[0], v1[1], c="purple", marker="o", label='Дом idx=257')
plt.scatter(v2[0], v2[1], c='forestgreen', marker='X', label='Дом idx=355')

plt.plot([0,v1[0]], [0, v1[1]], color="purple", marker=None, linewidth=0.8)
plt.plot([0,v2[0]], [0, v2[1]], color="forestgreen", marker=None, linewidth=0.8)

plt.title('Два дома как вектора в выбранном 2D пространстве (DIS, PTRATIO)')
plt.xlabel('DIS')
plt.ylabel('PTRATIO')
plt.grid(True)
plt.legend()
```

```python
# Выводы:
# Угол между векторами острый, значит скалярное произведение > 0. Существенно (почти в 2 раза) различаются длины векторов.
# Данные дома в 2D пространстве (DIS, PTRATIO) расположены далеко друг от друга, видны различия по обоим атрибутам.
```

Найти косинус угла между векторами x1 и x2, где x1 и x2 - это два дома из 2ух атрибутов, полученных на этапе 2.

```python
v1Length = np.linalg.norm(v1) # или так np.sqrt(np.dot(v1,v1))
v2Length = np.linalg.norm(v2)

cosine = np.dot(v1,v2)/(v1Length*v2Length)
cosine
```

```python
# угол между векторами в градусах
np.arccos(cosine) * 180/np.pi
```

Найти расстояние между точками x1 и x2.

```python
distance = math.hypot(v2[0] - v1[0], v2[1] - v1[1])
distance
```

# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
rcParams['figure.figsize'] = 6,6

new_X = x+300
new_Y = y+200

plt.plot(x, y, label='исходный объект')
plt.plot(new_X, new_Y, color='darkblue', label='после смещения')

plt.title('Смещение объекта: по x +300, по y +200')
plt.grid(True)
plt.legend()

plt.xlim([-200, 500])
plt.ylim([-200, 500])
```

Уменьшить объект в два раза.

```python
TransformArray = np.array([
    [0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
])
```

```python
B = []
for point in A:
    new_point=np.dot(TransformArray,np.transpose(point))
    B.append(new_point)

B = np.array(B).astype(int)
print(B)
```

```python
new_X = B[:,0]
new_Y = B[:,1]

plt.plot(x, y, label='исходный объект')
plt.plot(new_X, new_Y, color='red', label='после масштабирования')

plt.title('Уменьшение объекта в два раза')
plt.grid(True)
plt.legend()

plt.xlim([-200, 200])
plt.ylim([-200, 200])
```

Повернуть на 130 градусов.

```python
theta = 130 * math.pi/180 # угол в радианах

TransformArray = np.array([
    [math.cos(theta), math.sin(theta), 0],
    [-math.sin(theta), math.cos(theta), 0],
    [0, 0, 1]
])
```

```python
C = []
for point in A:
    new_point=np.dot(point,TransformArray)
    C.append(new_point)

C = np.array(C).astype(int)
print(C)
```

```python
new_X = C[:,0]
new_Y = C[:,1]

plt.plot(x, y, label='исходный объект')
plt.plot(new_X, new_Y, color='red', label='после поворота')

plt.title('Поворот объекта на 130 градусов против часовой стрелки')
plt.grid(True)
plt.legend()

plt.xlim([-200, 200])
plt.ylim([-200, 200])
```

Отразить объект относительно прямой y=x

```python
TransformArray = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
])
```

```python
K = []
for point in A:
    new_point=np.dot(point,TransformArray)
    K.append(new_point)

K = np.array(K).astype(int)
print(K)
```

```python
new_X = K[:,0]
new_Y = K[:,1]

plt.plot(x, y, label='исходный объект')
plt.plot(new_X, new_Y, color='darkblue', label='после отражения')

plt.title('Отражение объекта относительно прямой y=x')
plt.grid(True)
plt.legend()

plt.xlim([-200, 200])
plt.ylim([-200, 200])
```

Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```python
M1 = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [200, 300, 1]
])

M2 = np.array([
    [0.5, 0, 0],
    [0, 1.2, 0],
    [0, 0, 1]
])
```

```python
TransformArray = np.dot(M1, M2)
TransformArray
```

```python
R = []
for point in A:
    new_point=np.dot(point,TransformArray)
    R.append(new_point)

R = np.array(R).astype(float)
print(R)
```

```python
new_X = R[:,0]
new_Y = R[:,1]

plt.plot(x, y, label='исходный объект')
plt.plot(new_X, new_Y, color='darkblue', label='после преобразования')

plt.title('Cмещение объекта и масштабирование')
plt.grid(True)
plt.legend()

plt.xlim([-200, 500])
plt.ylim([-200, 500])
```
